```python
import pandas as pd
import numpy as np
df=pd.read_csv('부상 데이터셋.csv', encoding='UTF-8')
df
```

```python
df.dropna(inplace = True)
df['Incident Cost'] = df['Incident Cost'].str.replace(pat =r'[^\w]', repl=r'', regex=True ) #특수문자 제거

import numpy as np
bins = [0,10,20,30,40,50]
df['Age_grouping'] = np.digitize(df['Age'], bins)

df
```

```python
df['Injury Location'].unique()
```

```python
#Injury Location 묶어서 라벨링

map_Injury = {'Trunk' : 0, 'Abdomen' : 0, 'Back' : 0, 'Legs' : 1, 'Neck':2, 'Head' : 2, 'Multiple':3,
        'Eye':4, 'Arms': 5, 'Hands' : 5, 'Feet' : 1}
df['Injury Location'] = df['Injury Location'].map(map_Injury)
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['scaled_DaysLost','scaled_IncidentCost']] = scaler.fit_transform(df[['Days Lost','Incident Cost']])

df
```

```python
df.drop(['Age','Days Lost','Incident Cost', 'Incident Type','Date','ID','Unnamed: 5'], axis = 1, inplace = True)
         #시계열 데이터 :  시간에 따른 연속성
```

```python
col_names = ['Gender', 'Shift','Injury Location', 'Age_grouping', 'scaled_DaysLost','scaled_IncidentCost']
df.columns = col_names
```

```python
# 라벨링 처리 : 범주형 데이터를 수치형 데이터로 변환
df.loc[df['Gender']== 'Male','Gender'] = 0 #Gender열의 Male 데이터를 0으로 표시
df.loc[df['Gender']== 'Female','Gender'] = 1 #Gender열의 Female 데이터를 1으로 표시
df.loc[df['Shift']== 'Day','Shift'] = 0 #Shift열의 Day 데이터를 0으로 표시
df.loc[df['Shift']== 'Night','Shift'] = 1 #Shift열의 Night 데이터를 1으로 표시
df
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split

X=df[['Gender', 'Shift', 'Age_grouping', 'scaled_DaysLost','scaled_IncidentCost']].astype(float)
Y = df[['Injury Location']].astype(float)
Y=np.ravel(Y)
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state =1)
X
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

KC = KNeighborsClassifier(n_neighbors = 377)

KC.fit(X_train, y_train)
y_pred = KC.predict(X_test)
print('Accuracy: ', accuracy_score(y_test,y_pred))
```
